```python
def find_rms(l):
    for k in l:
        k.draw()
        points = k.get_axis_scaled_points()
        original_point = []
        calculated_point = []
        for j in points:
            #print(j)
            a = kmAxis.find_axis_value_at_point(list(j))
            #print(a)
            b = kmAxis.find_axis_point(a)
            original_point.append(j)
            calculated_point.append(b)
        
        squared_diffs = [(p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 for p1, p2 in zip(original_point, calculated_point)]
        mean_squared_diff = np.mean(squared_diffs)
        rmse = np.sqrt(mean_squared_diff)
        print(str(k), "RMSE:", rmse)
        
```

```python
for i in range(10):
    find_rms(bmi_n)
    find_rms(tutorial)
```
