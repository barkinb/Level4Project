```python
from matplotlib import pyplot as plt

height_numbers = [4, 6, 4, 4, 4, 4, 4, 4, 4]
bmi_numbers = [4, 6, 5, 4, 4, 4, 4, 4, 4]
mass_numbers = [4 ,4 ,4 ,4 ,4 ,4, 4, 4 ,4]

bins = [3.5, 4.5, 5.5, 6.5, 7.5]
plt.hist(height_numbers, bins=bins, color='skyblue', edgecolor='black', label='Height')
plt.hist(bmi_numbers, bins=bins, color='green', edgecolor='black', alpha=0.5, label='BMI')
plt.hist(mass_numbers, bins=bins, color='red', edgecolor='black', alpha=0.5, label='Mass')

plt.legend(loc='upper right')

plt.xlabel('Number of axis points')
plt.ylabel('Frequency')
plt.title('Number of Axis Points per Axis in the BMI Nomogram') 
plt.axvline(x=4.5, color='red', linestyle='--')
plt.xticks(range(4, 8))  
plt.yticks(range(1, 10))

plt.savefig('axis_histogram.pdf')
plt.show()
```

```python
from matplotlib import pyplot as plt

height_numbers = [4, 6, 4, 4, 4, 4, 4, 4, 4]
bmi_numbers = [4, 6, 5, 4, 4, 4, 4, 4, 4]
mass_numbers = [4 ,4 ,4 ,4 ,4 ,4, 4, 4 ,4]

bins = [3.5, 4.5, 5.5, 6.5, 7.5]
width = 0.3  # Width of each bar
plt.hist(height_numbers, bins=bins, align='left', color='skyblue', edgecolor='black', label='Height', width=width)
plt.hist(bmi_numbers, bins=bins, align='mid', color='green', edgecolor='black', label='BMI', width=width)
plt.hist(mass_numbers, bins=bins, align='right', color='red', edgecolor='black', label='Mass', width=width)

plt.legend(loc='upper right')

plt.xlabel('Number of axis points')
plt.ylabel('Frequency')
plt.title('Number of Axis Points per Axis in the BMI Nomogram') 
plt.axvline(x=4.5, color='red', linestyle='--')
plt.xticks(range(4, 8))  
plt.yticks(range(1, 10))

plt.savefig('axis_histogram.pdf')
plt.show()
```
