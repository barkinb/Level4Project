```python
import random
```

```python

def find_rms(axis):
    y = []
    original_axis_type = axis.get_axis_equation_type()
    original_points = axis.get_axis_scaled_points()
    original_values = []
    for i in range(len(original_points)):
        original_values.append(axis.find_axis_value_at_point(original_points[i]))
    new_values = []
    for z in range(10):
        new_axis_points = []
        for j in range(4):
            
            number = random.randrange(0,len(original_points))
            x = list(original_points[number])
            x.append(axis.find_axis_value_at_point(original_points[number]))
            new_axis_points.append(x)
        axis.set_axis_points(new_axis_points)
        for j in range(len(original_points)):
            new_values.append(axis.find_axis_value_at_point(original_points[j]))
        squared_diffs = []
        for k in range(len(original_values)):
            squared_diffs.append((original_values[k] - new_values[k])**2)
        mean_squared_diff = np.mean(squared_diffs)
        rmse = np.sqrt(mean_squared_diff)
        y.append(rmse)
    
    return y
            
```

```python
print(np.mean(find_rms(kmAxis)))
print(np.mean(find_rms(litres)))
print(np.mean(find_rms(kmsper)))
```
